```python
import requests

# Function to download file from URL
def download_file(url, local_filename):
    with requests.get(url, stream=True) as r:
        r.raise_for_status()
        with open(local_filename, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192):
                f.write(chunk)
    return local_filename

# List of file parts to be appended to the base URL
file_parts = [
    'nina34.anom.data',
    'meiv2.data'
]

# Base URL
base_url = 'https://psl.noaa.gov/data/correlation/'

# NOAA Monthly Climate Index Data Retrieval
for part in file_parts:
    url = base_url + part  # Create the full URL
    local_filename = part  # Use the file part as the local filename
    download_file(url, local_filename)  # Download the file
    shutil.move(local_filename, f'/content/drive/MyDrive/Colab Notebooks/WCDL/{local_filename}')  # Move the file to Google Drive
```
