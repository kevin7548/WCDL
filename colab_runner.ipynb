```python
# 2. Mount Google Drive (for data stored on Drive)
from google.colab import drive
drive.mount('/content/drive')

# Symlink data directory from Drive to repo
# Adjust the path below to match your Drive folder
import os
os.environ['WCDL_ROOT'] = '/content/WCDL'

DRIVE_DATA = '/content/drive/MyDrive/WCDL_data'  # <-- adjust this
REPO_DATA = '/content/WCDL/data'

if os.path.exists(DRIVE_DATA):
    if os.path.islink(REPO_DATA) or os.path.isdir(REPO_DATA):
        !rm -rf {REPO_DATA}
    os.symlink(DRIVE_DATA, REPO_DATA)
    print(f'Linked {DRIVE_DATA} -> {REPO_DATA}')
else:
    print(f'Drive path not found: {DRIVE_DATA}')
    print('Using local data/ directory instead.')
```
