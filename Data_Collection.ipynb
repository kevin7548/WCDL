Climate Indices

```python
# Function to download file from URL
def download_file(url, local_filename):
    with requests.get(url, stream=True) as r:
        r.raise_for_status()
        with open(local_filename, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192):
                f.write(chunk)
    return local_filename

# List of file parts to be appended to the base URL
file_names = ['aao', 'amo', 'amon.us.long', 'ao', 'censo.long', 'ea', 'espi', 'epo', 'gmsst', 'meiv2', 'nao', 'nina1.anom', 'nina3.anom', 'nina34.anom', 'nina4.anom', 'oni', 'pacwarm', 'pdo', 'pna', 'qbo', 'soi', 'solar', 'tna', 'tni', 'tsa', 'whwp', 'wp']

## IOD(Indian Ocean Dipole, dmi.long), PMM(Pacific Meridional Mode, pmmwind): do not exist

# Base URL
base_url = 'https://psl.noaa.gov/data/correlation/'

# NOAA Monthly Climate Index Data Retrieval
for file_name in file_names:
    url = base_url + file_name + '.data'  # Create the full URL
    local_filename = file_name + '.data'  # Use the file part as the local filename
    download_file(url, local_filename)  # Download the file
    shutil.move(local_filename, f'/content/drive/MyDrive/Colab Notebooks/WCDL/{local_filename}')  # Move the file to Google Drive

# Download ammsst.data from a different URL
ammsst_url = 'https://psl.noaa.gov/data/timeseries/monthly/AMM/ammsst.data'
ammsst_local_filename = 'ammsst.data'
download_file(ammsst_url, ammsst_local_filename)
shutil.move(ammsst_local_filename, f'/content/drive/MyDrive/Colab Notebooks/WCDL/{ammsst_local_filename}')
```

```python
dataset = "reanalysis-era5-single-levels-monthly-means"
request = {
    "product_type": ["monthly_averaged_reanalysis"],
    "variable": ["2m_temperature"],
    "year": [str(x) for x in range(1979, 2024)],
    "month": [str(x) for x in range(1, 13)],
    "time": ["00:00"],
    "data_format": "netcdf",
    "download_format": "unarchived",
    "area": [90, -180, -90, 180]
}

client = cdsapi.Client()
client.retrieve(dataset, request).download()
```

```python
# Initialize CDS API client
c = cdsapi.Client()

# Data collection and saving to Google Drive
for u, v in zip(['mean_sea_level_pressure', 'sea_ice_cover', 'mean_top_net_long_wave_radiation_flux'], ['mslp', 'sic', 'mtnlwr']):
    nc_filename = v + '.nc'

    # Retrieve data and save as netcdf file
    c.retrieve(
        'reanalysis-era5-single-levels-monthly-means',
        {
            'format': 'netcdf',
            'product_type': 'monthly_averaged_reanalysis',
            'variable': u,
            'year': [str(x) for x in range(1979, 2023)],
            'month': [str(x) for x in range(1, 13)],
            'time': '00:00',
            'grid': '2.5/2.5',
            'area': [90, -180, -90, 180],
        },
        nc_filename)

    # Move the file to Google Drive
    shutil.move(nc_filename, f'/content/drive/MyDrive/Colab Notebooks/WCDL/{nc_filename}')
```
